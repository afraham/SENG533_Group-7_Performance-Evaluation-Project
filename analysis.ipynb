Loading CSV Files

```python
import pandas as pd

colab = pd.read_csv("Benchmark Results/collab_svhn_benchmark_results.csv")
kaggle = pd.read_csv("Benchmark Results/kaggle_svhn_benchmark_results.csv")

colab["platform"] = "Colab"
kaggle["platform"] = "Kaggle"

df = pd.concat([colab, kaggle], ignore_index=True)
```

Graph 1: Platform Comparison

```python
import matplotlib.pyplot as plt

platform_avg = df.groupby("platform")["total_train_time_s"].mean()
platform_avg.plot(kind="bar")
plt.title("Average Training Time by Platform (Colab vs Kaggle)")
plt.ylabel("Training Time (seconds)")
plt.xticks(rotation=0)
plt.savefig("platform_comparison.png", bbox_inches='tight')
plt.show()
```

Graph 2: CPU vs GPU

```python
device_avg = df.groupby("device_kind")["total_train_time_s"].mean()
device_avg.plot(kind="bar")
plt.title("CPU vs GPU Training Time")
plt.ylabel("Training Time (seconds)")
plt.xticks(rotation=0)
plt.savefig("cpu_vs_gpu_comparison.png", bbox_inches='tight')
plt.show()
```

Graph 3: Model Comparison

```python
model_avg = df.groupby("model")["total_train_time_s"].mean()
model_avg.plot(kind="bar")
plt.title("Training Time by Model Type")
plt.ylabel("Training Time (seconds)")
plt.xticks(rotation=0)
plt.savefig("model_comparison.png", bbox_inches='tight')
plt.show()
```

Graph 4: Dataset Size Effect

```python
dataset_avg = df.groupby("dataset_fraction")["total_train_time_s"].mean()
dataset_avg.plot(kind="bar")
plt.title("Training Time: 50% vs 100% Dataset")
plt.ylabel("Training Time (seconds)")
plt.xticks(rotation=0)
plt.savefig("dataset_size_comparison.png", bbox_inches='tight')
plt.show()
```
